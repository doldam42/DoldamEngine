```python
import os
import glob

files = glob.glob('./**/*.h', recursive=True)
classes = []
count = 0
for f in files:
    dirname, filename = os.path.split(f)
    if dirname.find("packages") != -1:
        continue
    count+=1
    classes.append(filename)
    print(dirname)
classes = set(classes)
```
